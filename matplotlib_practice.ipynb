```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan","Feb","Mar","Apr","May","June","July","Aug","Sep","Oct","Nov","Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02,23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax=fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis,label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0,45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan","Feb","Mar","Apr","May","June","July","Aug","Sep","Oct","Nov","Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02,23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]

plt.scatter(y_axis, x_axis,c="red",label="Chicago")
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
```

```python
plt.scatter(x_axis,y_axis,s=y_axis)
```

```python
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
# Making the bubbles bigger using Python's list comprehension technique.
plt.scatter(x_axis,y_axis,s=[i * 3 for i in y_axis])
```

```python
# Create a scatter plot using object-orientged interface.
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
# Create a bubble chart using object-orientged interface.
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis,s=y_axis)
```

```python
# Create a bubble chart using object-orientged interface.
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis,c="skyblue", alpha = 0.2, s=y_axis, edgecolor="black", label="Boston")
plt.legend()
plt.title("Fare($) across months.")
plt.gca().invert_yaxis()
plt.xlim(0,50)
```

```python
# Create a pie chart.
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values=( 0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct="%.1f%%")
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue","magenta","lightblue","green","yellowgreen","greenyellow","yellow","orange","gold","indianred","tomato","mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,explode=explode_values,colors=colors,labels=x_axis,autopct="%.1f%%")
plt.show()
```

```python
# Create a pie chart using Object-oriented interface.
fig, ax = plt.subplots()
colors = ["slateblue","magenta","lightblue","green","yellowgreen","greenyellow","yellow","orange","gold","indianred","tomato","mistyrose"]
explode_values = (0,0,20,0,0,0,20,0,0,0,0,0)
ax.pie(y_axis, labels=x_axis, explode=explode_values, counterclock=False, colors=colors, radius=8*8,pctdistance=0.8, autopct="%0.1f%%",shadow=True,startangle=90)
ax.axis('equal')
plt.show()
```
